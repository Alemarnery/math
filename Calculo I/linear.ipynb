```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #Do make chart
```

```python
N = 100

y2= -5
y1=5
x2= -10
x1=10

m= (y2-y1/x2-x1)
b = 5

def f(x):
    return m*x + b


#Obteniendo mis coordenadas en X
x = np.linspace(-10,10, num=N)   #Obtengo 100 numeros en X desde [10 a -10]
y = f(x)

fig, ax = plt.subplots()
# Dibujar lneas para los ejes X y Y en el origen (0,0)
ax.axhline(y=0, color='black', linewidth=1)  # Eje X
ax.axvline(x=0, color='black', linewidth=1)  # Eje Y
plt.xlabel('x')  # Etiqueta para el eje X
plt.ylabel('y')  # Etiqueta para el eje Y
ax.plot(x,y)
ax.grid()
```
